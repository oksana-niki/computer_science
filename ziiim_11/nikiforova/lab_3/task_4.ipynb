# 3.3.4. Задача 4: "Решение дифференциальных уравнений"

### 1. Решите численно дифференциальные уравнения 1 и уравнение 2 Таблицы 3 по-отдельности на диапазоне [-100, 100], используя пакет scipy, взяв начальные условия из варианта.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Дифференциальные уравнения по варианту 1
# y'' - y' = 0 -> система: y1' = y2, y2' = y2
# y'' - y' = sin(x) -> система: y1' = y2, y2' = y2 + sin(x)

# Общее преобразование второго порядка к системе первого
def system1(y, x):
    y1, y2 = y
    dy1dx = y2
    dy2dx = y2
    return [dy1dx, dy2dx]

def system2(y, x):
    y1, y2 = y
    dy1dx = y2
    dy2dx = y2 + np.sin(x)
    return [dy1dx, dy2dx]

# Начальные условия
y0 = [3, 1]

# Диапазон
x = np.linspace(-100, 100, 1000)

# Решения
sol1 = odeint(system1, y0, x)
sol2 = odeint(system2, y0, x)

# Построение графика
plt.figure(figsize=(12, 6))
plt.plot(x, sol1[:, 0], label="Уравнение 1: $y'' - y' = 0$", color='blue')
plt.plot(x, sol2[:, 0], label="Уравнение 2: $y'' - y' = \sin(x)$", color='red')
plt.title("Решения дифференциальных уравнений")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid(True)
plt.show()
```

### 2. Решите систему уравнений 2 из Таблицы 1 для своего варианта (обычные дифференциальные уравнения) численно для диапазона [-100, 100] или покажите, что решения не существует. В решении понадобится пакет scipy.integrate.ode.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Задача: решить систему
# x' = 2x + y
# y' = 3x + 4y
# x(0) = 0.3, y(0) = 0.05
# диапазон: [-100, 100]

def system(t, u):
    x, y = u
    dxdt = 2 * x + y
    dydt = 3 * x + 4 * y
    return [dxdt, dydt]

# Диапазон времени и начальные условия
t_span = (-100, 100)
t_eval = np.linspace(-100, 100, 1000)
u0 = [0.3, 0.05]

# Решение
sol = solve_ivp(system, t_span, u0, t_eval=t_eval)

# Построение графика
plt.plot(sol.t, sol.y[0], label="x(t)")
plt.plot(sol.t, sol.y[1], label="y(t)")
plt.xlabel("t")
plt.ylabel("x(t), y(t)")
plt.title("Решение системы ODE")
plt.legend()
plt.grid(True)
plt.show()
```
