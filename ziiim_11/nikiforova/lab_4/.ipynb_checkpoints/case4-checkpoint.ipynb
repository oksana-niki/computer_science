# 3.4.4. Задача 4: "Преобразование уравнений в функции Python"

```python
from sympy import *
init_printing(use_unicode=True, wrap_line=True, no_global=True)

W, X, Y, Z = symbols('W X Y Z')
eq = X - (Z - X)/W
lam_eq = lambdify((W, X, Y, Z), eq, 'numpy')
print(lam_eq(-1/3, 12, 4, 8))
```

```python
my_sphere = lambda x, y, z, R: pow(x, 2) + pow(y, 2) + pow(z, 2) < pow(R, 2)

print(my_sphere(0.1, 0.1, 0.1, 1))   # True
print(my_sphere(1, 23, 2, 1))        # False
```

### 1. Преобразуйте выражение

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x, y, z: (
    np.pi * (x - 10025 * np.sin(x)) / y +
    1000 * np.pi + z -
    np.sqrt((x**4 - x**2 + 1) / (x**(1/4) + x**2 - 1))
)

x = np.linspace(0.1, 10, 500)  # избегаем деления на 0
y = 1
z = 0

plt.figure(figsize=(10, 6))
plt.plot(x, f(x, y, z), label='f(x)')
plt.title("График функции f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

### 2. Преобразуйте это же выражение в проверочное, выдающее «False», если результат меньше нуля и «True», если результат больше нуля.

```python
from math import pi, sin, pow

# Выражение:
# f(x, y, z) = pi * (x - 10025 * sin(x)) / y + 1000 * pi + z - sqrt((x**4 - x**2 + 1)/(x**(1/4) + x**2 - 1))

check_expr = lambda x, y, z: (
    pi * (x - 10025 * sin(x)) / y + 1000 * pi + z 
    - ((x**4 - x**2 + 1)/(x**(1/4) + x**2 - 1))**0.5
) > 0

# Пример использования
print(check_expr(1, 1, 1))  # True или False
```
