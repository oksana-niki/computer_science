# Построение линейного графика по двум массивам

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 100.0, 0.01)  # 1000 последовательных точек в пределах от 0 до 100
print(t[:10])                    # Печатаем первые 10 точек времени
y = np.exp(-t / 50) * (0.2 * np.sin(2 * t) + 0.3 * np.sin(0.9 * t))  # Вычисляем точки для функции
print(y[:10])                    # Печатаем первые 10 точек функции

fig, axis = plt.subplots()      # Создаем независимые фигуру и оси для нее
axis.plot(t, y, color='b')      # Создаем график y(x) синего (b-lue) цвета
axis.set(xlabel='time (s)', ylabel='voltage (mV)',
         title='Sample graph y(t)')  # Задаем подписи для осей и подпись графика
axis.grid()                     # Добавляем сетку
fig.savefig('saves/sample1_Lesson4.png')  # Сохраняем фигуру в файле
plt.show()                      # Отображаем график
```

# Построение двух графиков на одних осях

```python
t = np.arange(0.0, 100.0, 0.01)     # 1000 последовательных точек в пределах от 0 до 100
print(t[:10])                       # Печатаем первые 10 точек времени

y1 = np.exp(-t / 50) * (0.2 * np.sin(2 * t) + 0.3 * np.sin(0.9 * t))  # Вычисляем точки для функции 1
y2 = (0.2 * np.cos(2 * t) + 0.3 * np.cos(0.9 * t))                    # Вычисляем точки для функции 2
print(y1[:10])                     # Печатаем первые 10 точек функции

fig, axis = plt.subplots()        # Создаем независимые фигуру и оси для нее
axis.plot(t, y1, color='b', label='Функция 1')  # Создаем график y1(x) синего (b-blue) цвета
axis.plot(t, y2, color='r', label='Функция 2')  # Создаем график y2(x) красного (r-red) цвета

axis.set(xlabel='time (s)', ylabel='voltage (mV)',
         title='Sample graph y(t)')              # Задаем подписи для осей и подпись графика
axis.grid()                                      # Добавляем сетку
axis.legend()                                    # Добавляем легенду
fig.savefig('saves/sample2_Lesson4.png')         # Сохраняем фигуру в файл
plt.show()                                       # Отображаем график
```

# Построение графиков в разных осях на одной фигуре

# Построение различных видов графиков

```python
import mpl_toolkits.axes_grid1.axes_size as Size
from mpl_toolkits.axes_grid1 import Divider  # Дополнительные импорты

t = np.arange(0.0, 100.0, 0.01)  # 1000 последовательных точек в пределах от 0 до 100
r = np.exp(-t / 0.05)  # "импульсная характеристика"

# Для гистограмм
x = np.random.rand(len(t))
x = np.convolve(x, r)[:len(x)] * 0.01  # гауссовый ("цветной") шум

# Для pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']  # подписи
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # вытаскиваем 'Hogs'

fig, axis = plt.subplots(2, 3)  # Создаем независимые фигуры и оси для нее: 2 по вертикали и 3 по горизонтали
# fig.set_dpi(100)
fig.set_figheight(10)
fig.set_figwidth(10)

axis[0, 0].plot(t, np.sin(0.1*t), color='b')  # Создаем график синего (b-blue) цвета на осях с координатами (0,0)
axis[0, 1].scatter(t[:10], r[:10], color='r')  # Создаем точечный график красного (r-red) цвета на осях с координатами (0,1)
axis[0, 2].hist(x, 400, density=True, color='g')  # Создаем гистограмму зеленого (g-green) цвета на осях (0,2)
axis[1, 0].fill(t, r, color='b')  # Создаем график синего (b-blue) цвета на осях с координатами (1,0)

axis[1, 1].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
               shadow=True, startangle=90)  # Создаем "pie chart" на осях с координатами (1,1)
axis[1, 1].axis('equal')  # задаем жесткий размер осей (квадратная область, чтобы pie не превратился в эллипс)

axis[1, 2].plot(t, np.sin(t), color='g')  # Создаем график синего (g-green) цвета на осях с координатами (1,2)
axis[1, 2].grid(True)

axis[0, 0].set(xlabel='time (s)', ylabel='voltage (mV)',
              title='Линейный график')  # Задаем подписи для осей с координатами (0,0) и подпись графика
axis[0, 1].set(xlabel='time (s)', ylabel='voltage (mV)',
              title='Точечный график')  # Задаем подписи для осей с координатами (0,1) и подпись графика
axis[0, 2].set(xlabel='time (s)', ylabel='voltage (mV)',
              title='Гистограмма')  # Задаем подписи для осей с координатами (0,2) и подпись графика
axis[1, 0].set(xlabel='time (s)', ylabel='voltage (mV)',
              title='Область')  # Задаем подписи для осей с координатами (1,0) и подпись графика
axis[1, 1].set_title('Pie chart')  # Подпись графика
axis[1, 2].set_title('Синус')  # Подпись графика

fig.tight_layout()  # Предотвращаем наложение осей друг на друга
fig.savefig('saves/sample6_Lesson3.png')  # Сохраняем фигуру в файл с большим разрешением
plt.show()  # Отображаем график
```

# Построение графика функции двух переменных

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
from scipy.stats import multivariate_normal

# шаг сетки
delta = 0.5
x = np.arange(-3.0, 4.001, delta)
y = np.arange(-4.0, 3.001, delta)
X, Y = np.meshgrid(x, y)

# заменяем bivariate_normal с помощью scipy
pos = np.dstack((X, Y))
Z1 = multivariate_normal(mean=[0, 0], cov=[[1.0, 0], [0, 1.0]]).pdf(pos)
Z2 = multivariate_normal(mean=[1, 1], cov=[[1.5, 0], [0, 0.5]]).pdf(pos)
Z = Z1 - Z2

# уровни
levels = np.linspace(-0.1, 0.1, 40)
norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())

fig = plt.figure(figsize=plt.figaspect(0.5))
fig.set_dpi(140)

# оси
axis2d = fig.add_subplot(1, 2, 1)
axis3d = fig.add_subplot(1, 2, 2, projection='3d')

# 2D контур
cset1 = axis2d.contourf(X, Y, Z, levels=levels, norm=norm)

# освещение
ls = LightSource(270, 45)
rgb = ls.shade(Z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft')

# 3D поверхность
surf = axis3d.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=rgb,
                           linewidth=0, antialiased=False, shade=False)

# легенды
fig.colorbar(cset1, shrink=0.5, aspect=10)
fig.colorbar(surf, shrink=0.5, aspect=10)

# сохранить и показать
fig.savefig('saves/sample6_Lesson3.png', dpi=300)
plt.show()
```

# Построение вращающегося трёхмерного графика с лимитированным временем жизни и графическим интерфейсом

```python
'''
====================
Rotating a 3D plot
====================

A very simple animation of a rotating 3D plot.

See wire3d_animation_demo for another simple example of animating a 3D plot.
'''

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Load some test data for demonstration and plot a wireframe
X, Y, Z = axes3d.get_test_data(0.1)
ax.plot_wireframe(X, Y, Z, rstride=5, cstride=5)

# rotate the axes and update
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```

# Построение автоматического графопостроителя с консолью и графическим интерфейсом

```python
"""
=======
Textbox
=======

Allowing text input with the Textbox widget.

You can use the Textbox widget to let users provide any text that needs to be
displayed, including formulas. You can use a submit button to create plots
with the given input.
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import TextBox

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.2)

t = np.arange(-2.0, 2.0, 0.001)
s = t ** 2
initial_text = "t ** 2"
l, = plt.plot(t, s, lw=2)

def submit(text):
    ydata = eval(text)
    l.set_ydata(ydata)
    ax.set_ylim(np.min(ydata), np.max(ydata))
    plt.draw()

axbox = plt.axes([0.1, 0.05, 0.8, 0.075])
text_box = TextBox(axbox, 'Evaluate', initial=initial_text)
text_box.on_submit(submit)

plt.show()
```
