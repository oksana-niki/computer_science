# 3.5.1. Задача 1: "График по двум массивам"

### Рассчитайте Y по функциям 1 и 2 из таблицы 2, используя x в диапазоне [0, 100] и постройте их на одном графике.

```python
import numpy as np
import matplotlib.pyplot as plt

# Диапазон x
x = np.linspace(0, 100, 1000)

# Функция 1: y(x) = sin(3x)
y1 = np.sin(3 * x)

# Функция 2: y(x) = 2cos(x)
y2 = 2 * np.cos(x)

# Построение графиков
plt.plot(x, y1, label='y(x) = sin(3x)', color='b')
plt.plot(x, y2, label='y(x) = 2cos(x)', color='r')

# Настройки графика
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графики функций y1 и y2')
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```
