# 3.5.2. Задача 2: "Работа с множеством осей"

### Постройте те же функции 1 и 2 из первой задачи на разных осях, разместив оси друг под другом. Для этого воспользуйтесь командой subplots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаём массив x от 0 до 100
x = np.linspace(0, 100, 1000)

# Функции из таблицы 2
y1 = np.sin(3 * x)
y2 = 2 * np.cos(x)

# Создаём подграфики (2 строки, 1 столбец)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Первый график
ax1.plot(x, y1, label='y = sin(3x)', color='blue')
ax1.set_ylabel('y1')
ax1.legend()
ax1.grid(True)

# Второй график
ax2.plot(x, y2, label='y = 2cos(x)', color='green')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
