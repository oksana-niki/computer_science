На одном полотне (фигуре) постройте:

1. Постройте те же функции 1 и 2 из первой задачи на разных осях, разместив оси друг под другом;

2. Постройте гистограмму распределений 1 и 2, приведённых для вашего варианта в Таблице 2, использовав заданные параметры. Постройте две гистограммы на одном графике и две гистограммы на раздельных осях одной фигуры;

3. Добавьте на еще одни оси «Pie Chart» из теоретической части;

Задайте такие размеры полотна «figure», чтобы все графики отображались корректно.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Подготовка данных
x = np.linspace(0, 100, 1000)
y1 = np.sin(3 * x)
y2 = 2 * np.cos(x)

# Распределения
uniform_data = [random.uniform(-1, 1) for _ in range(1000)]
beta_data = [random.betavariate(2, 8) * 200 - 100 for _ in range(1000)]

# Данные для Pie Chart
sizes = [15, 35, 45]
labels = ['Frogs', 'Hogs', 'Dogs']
explode = (0, 0.1, 0)

# Создание фигуры
fig, axis = plt.subplots(2, 2, figsize=(12, 8))
fig.tight_layout(pad=4.0)

# Графики функций
axis[0, 0].plot(x, y1, label='y=sin(3x)')
axis[0, 0].set_title('y = sin(3x)')
axis[0, 1].plot(x, y2, label='y=2cos(x)', color='r')
axis[0, 1].set_title('y = 2cos(x)')

# Гистограммы
axis[1, 0].hist(uniform_data, bins=20, alpha=0.7, label='uniform a=-1, b=1')
axis[1, 0].hist(beta_data, bins=20, alpha=0.7, label='betavariate a=2, b=8')
axis[1, 0].set_title('Histograms')
axis[1, 0].legend()

# Pie chart
axis[1, 1].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
axis[1, 1].axis('equal')
axis[1, 1].set_title('Pie Chart')

plt.show()
```
