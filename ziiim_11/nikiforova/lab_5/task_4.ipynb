### Постройте график функции 3 из Таблицы 2, использовав двумерную сетку.

### Постройте контурный и трехмерный график.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

# Создаем двумерную сетку
x = np.arange(1, 11)
y = np.arange(1, 11)
X, Y = np.meshgrid(x, y)

# Вычисляем Z по формуле
Z = (factorial(X) - factorial(Y)) / factorial(X)

# Контурный график
plt.figure()
cp = plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(cp)
plt.title('Контурный график Z(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# Трехмерный график
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Трехмерный график Z(x, y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Z')
plt.show()
```
